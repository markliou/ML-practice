## Tokenrizers

```python
examples, metadata = tfds.load('ted_hrlr_translate/pt_to_en', with_info=True,as_supervised=True)
train_examples, val_examples = examples['train'], examples['validation']
```

```python
tr_iter = iter(train_examples)
pt, en = next(tr_iter)
print(pt)
print(pt.numpy())
print(pt.numpy().decode('utf-8'))
```

tokenizer主要有兩種:
1. Bert tokenizer: BERT製作時，會有特殊的表示字(e.g "[PAD]", "[UNK]", "[START]", "[END]")；同時是使用top-down的方式進行文字拆解
2. Piece word tokenizer:主要都建構在BPE演算法上，屬於bottom-up的算法
